```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for loop_type in ("train", "val"):
    for metric in ("loss", "acc"):
        df = pd.read_csv(f"{loop_type}_{metric}.csv")

        plt.figure()
        plt.plot(df["Step"], df["Value"])
        plt.xlabel("batch")
        plt.ylabel("value")
        plt.title(f"{loop_type} {metric}")
        plt.grid()
        plt.show()
```
